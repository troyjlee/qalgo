```python
import math
import numpy as np
from scipy import fftpack
from scipy.linalg import dft
import matplotlib.pyplot as plt
```

```python
N = 65
```

```python
s=10
```

```python
T = math.floor(N/s)
```

```python
R= N%s
```

```python
I = np.identity(s)
```

```python
X = np.tile(I,(T,1))
```

```python
X = np.vstack((X,I[0:R,:]))
```

```python
X = X/np.sqrt(sum(X))
```

```python
X.shape
```

```python
X.T@X
```

```python
F = dft(N)
```

```python
Y = F@X/np.sqrt(N)
```

```python
Y.conj().T@Y
```

```python
plt.plot(np.abs(Y[:,1]),'r')
plt.plot(np.abs(Y[:,2]),'b')
plt.ylabel('magnitude')
plt.show()
```
